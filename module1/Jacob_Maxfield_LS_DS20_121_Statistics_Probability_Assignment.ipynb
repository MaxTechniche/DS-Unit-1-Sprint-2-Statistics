````python
import pandas as pd
import numpy as np
````

````python
# Loading in Automobile Dataset
auto_df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data', header=None, na_values='?')
auto_df.head()
````

````python
column_names = ['symboling', 'normalized-losses', 'make', 'fuel-type', 
                'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 
                'engine-location', 'wheel-base', 'length', 'width', 'height', 
                'curb-weight', 'engine-type', 'num-of-cylinders', 'engine-size', 
                'fuel-system', 'bore', 'stroke', 'compression-ratio', 
                'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
````

````python
auto_df.columns = column_names
auto_df.head()
````

## 2) Pretend that this dataset represents the cars at a used car lot in your local town. 

### 2) part 1

- Use df.sample() to pick a random sample of 10 cars. Note that because this sample is random we are going to set the `random_state` so that all of us in the class get the same random sample. Please set your random state to `30` when using `df.sample()`

Is your sample reflective of the population value in regards to highway-mpg? Find the mean for `highway-mpg` for the entire dataset and use a 1-sample t-test to compare your estaimated sample mean to the population mean. Can you say that they are different? 

````python
from scipy import stats
````

````python
# Getting the mean of 'highway-mpg'
h_mpg_mean = auto_df['highway-mpg'].mean()
h_mpg_mean
````

**Actual highway-mpg ~ 30.75**

````python
# Generating a random sample
auto_sample = auto_df.sample(10, random_state=30)
auto_sample
````

### Null Hypothesis:

The average highway_mpg of the sample will be that of the average of the lots average.

### Alternative Hypothesis:

The sample highway_mpg mean is not the same as the lots average.

### Confidence Level:

**95% Confidence Level**

````python
# Performing a 1 sample ttest on the data
ttest_value = stats.ttest_1samp(auto_sample['highway-mpg'], h_mpg_mean)
ttest_value
````

### Conclusion:

Due to a t-statistic of .891 and a p-value of .396, I **fail to reject** the null hypothesis that the sample is a fair representation of the population highway-mpg mean.

### 2) part 2

The salesman says the cars he sells typically have a fuel efficiency of about 35 miles per gallon on the highway. You want to verify his claim but can only test 10 cars. Using your sample of 10, test his claim and report your results.

````python
# I already have my random sample
auto_sample
````

### Null Hypothesis:

The salesman's average of about 35 is the lots average.

### Alternative Hypothesis:

The salesman's average of about 35 is not the lots average.

### Confidence Level:

**90% Confidence Level**

````python
# Performing a 1 sample ttest on the sample vs the salesmans average
ttest_value = stats.ttest_1samp(auto_sample['highway-mpg'], 35)
ttest_value
````

### Conclusion

Due to a t-statistic of -.957 and a p-value of .364, I **fail to reject** the null hypothesis that the salesman's average of about 35 is the lots average.

### 2) Part 3

Imagine that you now have the capacity to test 100 cars on the lot. Using the same random state of `30`, take a sample of 100 cars. Run a t-test to verify the salesman's claim with your new larger sample. Do you reach the same conclusion as you did with the sample size of 10? 

````python
# Pulling a sample of 100 and making sure that I do in fact have 100 rows
large_sample = auto_df.sample(100, random_state=30)
large_sample.shape
````

````python
# Performing a 1 sample ttest on the larger sample vs the salesmans average
large_ttest_value = stats.ttest_1samp(large_sample['highway-mpg'], 35)
large_ttest_value
````

### Conclusion

Due to a t-statistic of -6.519 and a p-value of < .00001, I **reject** the null hypothesis that the salesman's given average (35) is the lots average.

## Practice 2-Sample T-tests

Two Sample t-tests determine whether or not two sample means are statistically different from each other. 

This portion of your assignment requires you to determine which issues have "statistically significant" differences between political parties in this [1980s congressional voting data](https://archive.ics.uci.edu/ml/datasets/Congressional+Voting+Records). 


### 1) Load the data

The data consists of 435 instances (one for each congressperson), a class (democrat or republican), and 16 binary attributes (yes or no for voting for or against certain issues). Be aware - there are missing values!

- Read the dataset in from UCI, you'll need to provide a list of column headers
- Encode "yes" votes as 1 and "no" votes as 0. (You can use `df.replace()` to swap out these values)
- Use dataframe filtering to split the dataframe into two new dataframes based on party. Hold all republicans in one dataframe and all democrats in the other. These will be our two different "samples."



````python
column_names = ['Class Name', 'handicapped-infants', 'water-project-cost-sharing', 
                'adoption-of-the-budget-resolution', 'physician-fee-freeze',
                'el-salvador-aid', 'religious-groups-in-schools', 'anti-satellite-test-ban',
                'aid-to-nicaraguan-contras', 'mx-missile', 'immigration', 
                'synfuels-corporation-cutback', 'education-spending', 'superfund-right-to-sue',
                'crime', 'duty-free-exports', 'export-administration-act-south-africa']
````

````python
# Loading in dataset using column_names
congress = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/voting-records/house-votes-84.data', names=column_names)
congress
````

````python
# Replacing all y's with 1, n's with 0, and ?'s with NaN
congress = congress.replace({'y': 1, 'n':0, '?':np.NaN, 'nan':np.NaN})
congress
````

````python
# Splitting the DataFrame into 2. (Those that are Republican and those that are Democrat)
rep = congress[congress['Class Name'] == 'republican']
dem = congress[congress['Class Name'] == 'democrat']
````

### 2) Perform two sample T-tests on different issues and report the results.

- Find an issue that democrats support more than republicans with p < 0.01 (significant at the 99% level).
- Find an issue that republicans support more than democrats with p < 0.01 (significant at the 99% level).
- Find an issue where the difference between republicans and democrats has p > 0.1 (Not significant at the 90% level - i.e. there may not be much of a difference the two sample means)

Please for each test that you run state your null and alternative hypothesis and then write a conclusion reflecting on the null and alternative hypothesis.

Remember, that two-sample t-tests will only tell us if the two groups are *different* from one another. We'll have to look at their sample means directly or use the sign on the t-statistic to know which of the two sample means is larger. 

````python
# I'm just going to do the for loop right now
for issue in congress.columns[1:]:
  test = stats.ttest_ind(dem[issue], rep[issue], nan_policy='omit')
  print(issue)
  print('Null Hypothesis: Democrats and Republicans have the same level of support on this bill.')
  print('Alternative Hypothesis: Democrats and Repubicans do not have the same level of support on this bill.')
  print(test)
  if test.pvalue < .01:
    reject = 'REJECT'
  else:
    reject = 'FAIL TO REJECT'
  print(f'Conclusion: Based on a t-statistic of about {test.statistic:.3f} and a p-value of about {test.pvalue:.3f},\n\t\
    we {reject} the null hypothesis that Democrats and Republicans have the same level of support on this bill.')
  if reject == 'REJECT':
    if test.statistic > 0:
      party = "Democrat's"
    else:
      party = "Republican's"
    print(f'\t{party} support it more.')
  print(f'\tDemocrat issue mean: {round(dem[issue].mean(), 2)}, Republican issue mean: {round(rep[issue].mean(), 2)}')
  print()
````

````python
# I'm just going to do the for loop right now
for issue in congress.columns[1:]:
  test = stats.ttest_ind(dem[issue], rep[issue], nan_policy='omit')
  print(issue)
  dem_var = sum([(x-dem[issue].mean())**2 for x in dem[issue] if x in [0, 1]])/dem[issue].count()
  rep_var = sum([(x-rep[issue].mean())**2 for x in rep[issue] if x in [0, 1]])/rep[issue].count()
  t_statistic = (dem[issue].mean()-rep[issue].mean())/math.sqrt(dem_var/dem[issue].count()+rep_var/rep[issue].count())
  print('Based off of aproximation (and continuous rounding), the absolute value of the t-statistic needs to be above about 2.6')
  print(f't-statistic of {t_statistic}')  
  print(test)
  if abs(t_statistic) > 2.6:
    reject = 'REJECT'
  else:
    reject = 'FAIL TO REJECT'
  print(f'Conclusion: Based on a t-statistic of about {t_statistic:.3f},\n\t\
    we {reject} the null hypothesis that Democrats and Republicans have the same level of support on this bill.')
  if reject == 'REJECT':
    if t_statistic > 0:
      party = "Democrat's"
    else:
      party = "Republican's"
    print(f'\t{party} support it more.')
  print(f'\tDemocrat issue mean: {round(dem[issue].mean(), 2)}, Republican issue mean: {round(rep[issue].mean(), 2)}')
  print()
````
